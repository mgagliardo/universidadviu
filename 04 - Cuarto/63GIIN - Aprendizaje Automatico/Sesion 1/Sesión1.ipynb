### Explicación paso a paso para crear un modelo de ML simple

```python
# Librerias
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar y dividir los datos
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=1)

# Crear y entrenar el modelo
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Evaluar el modelo
score = knn.score(X_test, y_test)
print(f'La precisión del modelo es del {score*100:.2f}%')

# Visualizar las características en pares
plt.figure(figsize=(12, 12))
for i in range(4):
    for j in range(4):
        plt.subplot(4, 4, i * 4 + j + 1)
        if i == j:
            plt.text(0.5, 0.5, iris.feature_names[i], ha='center', va='center')
            plt.xticks([])
            plt.yticks([])
        else:
            plt.scatter(X_train[:, j], X_train[:, i], c=y_train, cmap='viridis', marker='.')
            plt.xlabel(iris.feature_names[j])
            plt.ylabel(iris.feature_names[i])
plt.tight_layout()
plt.show()

# Visualizar la matriz de confusión
predictions = knn.predict(X_test)
conf_mat = confusion_matrix(y_test, predictions)

plt.figure(figsize=(6, 6))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion matrix')
plt.show()

```

**una matriz de confusión** es una herramienta que permite visualizar el desempeño de un algoritmo  de aprendizaje supervisado. Cada columna de la matriz representa el número de predicciones de cada clase, mientras que cada fila representa a las instancias en la clase real., o sea en términos prácticos nos permite ver  qué tipos de aciertos y errores está teniendo nuestro modelo a la hora de pasar por el proceso de aprendizaje con los datos.

```python
# Lo primero importar el módulo Numpy.
import numpy as np

# La función array permite crear un array de numpy a partir de una lista.
# Vamos a convertir una lista en un array:
list1 =[1,2,3,4]

# A partir de la lista cramos el array
array1= np.array(list1)

# otra opción para mostras sin print es poner el nombre del objeto: (Comprobar la diferencia de uno a otro)
#array1
print (array1)
list1
```

```python
# Este sencillo ejemplo permite ver como crear una array 2D.
list2 = [list1 , list1]
print (list2)

array2= np.array(list2)
#Con el método shape se puede determinar el tamaño del array.
print ('El tamaño del array es:\n', array2.shape )

#El método dtype podemos averiguar el tipo de datos contenido en el array.
print('El tipo de datos:\n',array2.dtype)

#Existen arrays especiales muy utilizados en ciencia de datos, por ejemplo la función zeros permite crear un array de ceros y la función ones permite crear un array de unos.
print ('Array de ceros:\n ', np.zeros((3,5)))
print ('Array de unos:\n', np.ones((5,3)))

#La función eye permite crear un array que contiene la matriz identidad del tamaño deseado.
print ('Matriz de identidad:\n' , np.eye(3))
```

Arrays y escalares
* Podemos hacer operaciones aritméticas con los elementos de un arrays.

```python
# El operador * multiplica los elementos de dos arrays elemento a elemento. formas diferentes
print('por numero:\n', array2 * 12)
print('por otor array:\n', array2 * array1)

#La resta se realiza elemento a elemento.
print('Menos constante:\n', array2 - 1)
print('Menos otro array:\n', array2 - array1)

#También podemos hacer operaciones que involucren arrays y escalares
# Si dividimos es una division escalar entre cada uno de los elementos del array
print('por numero:\n', 5/ array2)
print('por otor array:\n', array2 / array1)

#El operador de potencia es **, se realiza elemento a elemento
print('por numero:\n', array2 ** 4)
print('por otor array:\n', array2 ** array1)
```

Trasposición de arrays

```python
#Empecemos creando un array de 50 elemento y dándole forma, 10x5.
arr = np.arange(50).reshape((10,5))
print('Valor array : \n', arr)

# Para despues transponerla, agregando el atributo T
print('Valor array despues transponerla: \n', arr.T)

# Si queremos intercambiar los ejes del array también podemos utilizar el método swapaxes. Veamos un ejemplo.
print('Valor array despues intercambiar los ejes: \n',arr.swapaxes(0,1))

# Los arrays de NumPy pueden ser N-dimensionales, se crear un array de orden 3.
arr3d = np.arange(50).reshape((5,5,2))
print('Valor array N-dimensionales: \n', arr3d)

#Con el método transpose podemos transponer o intercambiar los ejes de la matriz a nuestro antojo. En este caso colocamos el eje 1 en el 0, el 0 en el 1 y
#el 2 lo dejamos en su sitio. El argumento es una tupla.
arrT= arr3d.transpose((1,0,2))
print('Valor array transpose: \n', arrT)
```

```python
# Importamos la libreria de pandas para poder trabajr con sus funciones
# Las principales funciones para crear las estructuras de datos de panda son Series y DataFrame.
import pandas as pd

# Una Serie en pandas constituye un array de unidimensional de objetos de cualquiera de los tipos que emplea NumPy.
# Esta estructura de datos se caracteriza porque permite almacenar el valor (value) y un índice (index) asociado a cada muestra
# obj = Series([4,7,5,3], index=['c','s','d','h'])
# print ('Crear un serie:\n',obj)

# Como habrás pensado a estas alturas, las Series de Pandas se parecen muchísimo a un diccionario de Python.
sdata = {'Sara': 23, 'Alex' : 31, 'Juan' : 22 , 'alba': 29}
obj1 = Series(sdata)
print ('Serie transformada:\n',obj1)

# Si utilizamos un diccionario para crear una serie, los indices de la serie coinciden con las claves del diccionario.
# Puesto que el índice ‘Andres’ no está contenido en el diccionario, NaN es asignado en su valor.
Alumnos = ['Andres','Sara','Alex','alba']
obj2 = Series(sdata, index=Alumnos)
print ('Cambiar valor de un diccionario:\n',obj2)

# Una DataFrame en pandas constituye una representación tabular (2D) con una colección de columnas que pueden ser de diferente tipo
# si la columna no está contenida en el diccionario, al igual que con las series, se incorpora al dataframe como NaN.
data = {'edad': [21,32,35,28,22], 'Trabajo':['local','remoto','local','local','remoto'], 'fecha':[2020,2012,2018,2018,2021]}
frame= DataFrame(data)
print(frame)
```

Normalizar el Dataframe de los Pandas con la media normalización


```python
#.apply() : Las funciones que mutan el objeto pasado.
# lamba : funcion pueden tener cualquier número de argumentos, pero solo una expresión
# .mean() : muestra la media aritmética del conjunto de datos proporcionado
# .std() :Devuelve la desviación estándar de la muestra sobre el eje solicitado

def mean_norm(df):
    return df.apply(lambda x: (x-x.mean())/ x.std(), axis=0)

df_mn = mean_norm(df)
df_mn
```

Aprovechamos esta serie para observar cómo eliminarlos Outliers estos son valores atípicos los cuales podrían afectar a la media de los datos.

```python
def outliers(df_in):
  low = 0.05
  high = 0.95
  quant_df  = df_in.quantile([low, high])
  df1 = df_in.apply(lambda x: x[(x > quant_df.loc[low,x.name]) & (x < quant_df.loc[high,x.name])], axis=0)
  return df1


df1 = outliers(df)
df1
```

```python
def expand_embark_acronym(embarked):
    result = []
    mapping = {
            "C": "Cherbourg",
            "S": "Southampton",
            "Q": "Queenstown"
    }
    for each in embarked.values:
        if len(str(each)) > 1:
            result.append(each)
        else:
            if each in mapping:
                result.append(mapping[each])
            else:
                result.append("Unknown")
    return result

def expand_pclass_acronym(pclass):
    result = []
    mapping = {
            1: "1st class",
            2: "2nd class",
            3: "3rd class"
    }
    for each in pclass.values:
        if len(str(each)) > 1:
            result.append(each)
        else:
            if each in mapping:
                result.append(mapping[each])
            else:
                result.append("Unknown")
    return result

def is_a_minor(age):

    if age < 18:
        return "Under 18 (minor)"

    return "Adult"

def apply_age_groups(age):
    result = []
    mapping = {
            1: "Infant",      # bebes menores de 1 años
           13: "Child",       # Menores de 12 años
           18: "Teen",        # 13-17 años
           66: "Adult",       # Adultos: +18
           123: "Elderly"     # Ancianos: +65
    }
    for each_age in age.values:
        if type(each_age) == str:
            result.append(category)
        else:
            category = "Unknown"
            if each_age != np.nan:
                for each_age_range in mapping:
                    if  each_age < each_age_range:
                        category = mapping[each_age_range]
                        break
            result.append(category)
    return result

def apply_age_ranges(age):
    result = []
    mapping = {
            6: "00-05 years",
           12: "06-11 years",
           19: "12-18 years",
           31: "19-30 years",
           41: "31-40 years",
           51: "41-50 years",
           61: "51-60 years",
           71: "61-70 years",
           81: "71-80 years",
           91: "81-90 years",
           124: "91+ years",  # (123 is the oldest age known till date)
    }

    for each_age in age.values:
        if type(each_age) == str:
            result.append(category)
        else:
            category = "Unknown"
            if each_age != np.nan:
                for each_age_range in mapping:
                    if  each_age < each_age_range:
                        category = mapping[each_age_range]
                        break
            result.append(category)
    return result

def is_married_of_single(names, ages, sexes):
    result = []
    for name, age, sex in zip(names.values, ages.values, sexes.values):
        if age < 18:
            result.append("Not of legal age")
        else:
            if ('Mrs.' in name) or ('Mme.' in name):
                result.append("Married")
            elif ('Miss.' in name) or ('Ms.' in name) or ('Lady' in name) or ('Mlle.' in name):
                result.append("Single")
            else:
                result.append("Unknown")

    return result

def apply_travel_companions(siblings_spouse, parent_children):
    result = []
    for siblings_spouse_count, parent_children_count in zip(siblings_spouse.values, parent_children.values):
        if (siblings_spouse_count > 0) and (parent_children_count > 0):
            result.append("Parent/Children & Sibling/Spouse")
        else:
            if (siblings_spouse_count > 0):
                result.append("Sibling/Spouse")
            elif (parent_children_count > 0):
                result.append("Parent/Children")
            else:
                result.append("Alone")

    return result

def apply_fare_ranges(fare):
    result = []
    mapping = {
           11: "£000 - 010",
           21: "£011 - 020",
           41: "£020 - 040",
           81: "£041 - 080",
          101: "£081 - 100",
          201: "£101 - 200",
          301: "£201 - 300",
          401: "£301 - 400",
          515: "£401 & above"  # in this case the max fare is around £512
    }
    for each_fare in fare.values:
        if type(each_fare) == str:
            result.append(category)
        else:
            category = "Unknown"
            if each_fare != np.nan:
                for each_fare_range in mapping:
                    if  each_fare < each_fare_range:
                        category = mapping[each_fare_range]
                        break
            result.append(category)

    return result

def were_in_a_cabin_or_not(row):
    if type(row) is str:
        return "En cabina"
    return "Fuera de la cabina"
```
